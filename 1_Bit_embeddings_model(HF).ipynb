```python
!pip install --upgrade langchain-community  # Upgrade to the latest version

from langchain.document_loaders import PyPDFLoader  # Update import path
from langchain_community.vectorstores import FAISS
from langchain_groq import ChatGroq
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains import create_retrieval_chain
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
loader = PyPDFLoader("/content/harry_porter_book.pdf")
docs = loader.load()
```

```python
docs
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
documents = text_splitter.split_documents(docs)
```

```python
len(documents)
```

```python
groq_api_key = "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"
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name = "mixedbread-ai/mxbai-embed-large-v1",
    encode_kwargs = {'precision' : 'binary'}
)
one_bit_vectorstore = FAISS.from_documents(documents, embeddings)
one_bit_retriever = one_bit_vectorstore.as_retriever(
    search_kwargs = {'k' : 3}
)
```

```python
prompt = ChatPromptTemplate.from_template("""
Answer the question  on the basis of the given context only. Don't add any extra information but just frame the answer nicely on the basis of given context.
context : {context}
question : {input}
""")
```

```python
model = ChatGroq(temperature=0, groq_api_key=groq_api_key, model_name="mixtral-8x7b-32768")
```

```python
document_chain = create_stuff_documents_chain(model, prompt)
one_bit_retrieval_chain = create_retrieval_chain(one_bit_retriever, document_chain)
```

```python
from operator import itemgetter
from langchain_core.runnables import  RunnablePassthrough
```

```python
chain = (
    {"context" : itemgetter("query") | one_bit_retriever ,"input" : itemgetter("query") }
    | RunnablePassthrough()
    | prompt
    | model
)
```

```python
%%time
response_1bit = chain.invoke({"query":"What was the character of Harry Porter ?"})
print(response_1bit)
```

# 32bits model
Now we would load 32 bits model to see its impact and time difference from binary model loaded above.
We would not prvide encode_keyword_arguments as "encode_kwargs = {'precision' : 'binary'}" which will load 32 bits model.

```python
embeddings = HuggingFaceEmbeddings(
    model_name = "mixedbread-ai/mxbai-embed-large-v1",

)
Thitytwo_bit_vectorstore = FAISS.from_documents(documents, embeddings)
Thitytwo_bit_retriever = Thitytwo_bit_vectorstore.as_retriever(
    search_kwargs = {'k' : 3}
)
```

```python
content32 = Thitytwo_bit_retriever.invoke("What was the character of Harry Porter ?")
content32
```

```python
content1 = one_bit_retriever.invoke("What was the character of Harry Porter ?")
content1
```

```python
def pretty_doc(content):
  list = []
  for i in range(3):
        doc = " ".join((content[i].page_content).split("\n")[1:])
        list.append(doc)
  doc_final =  " ".join(list)
  return doc_final

```

```python
one_bit_reference_doc = pretty_doc(content1)
thirtytwo_bit_reference_doc = pretty_doc(content32)
```

```python
print(one_bit_reference_doc)
```

```python
print(thirtytwo_bit_reference_doc)
```

```python
Thitytwo_bit_retriever.invoke("What was the character of Harry Porter ?")
```

```python
chain_thirty_two_bit = (
    {"context" : itemgetter("query") | Thitytwo_bit_retriever ,"input" : itemgetter("query") }
    | RunnablePassthrough()
    | prompt
    | model
)
```

```python
%%time
response_32bit = chain_thirty_two_bit.invoke({"query":"What was the character of Harry Porter ?"})
print(response_32bit)
```

```python
print(response_32bit.content)
```

```python
print(response_1bit.content)
```

Here wall time is 441ms for 32-bit embeddings model where as for 1-bit embeddings model its 385ms .This difference  will be more in case of large corpus.
